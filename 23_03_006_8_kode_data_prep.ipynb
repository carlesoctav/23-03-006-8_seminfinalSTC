```python
# B. Mempersiapkan Credit Score File
```

```python
cred_score= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/credit_record.csv')
cred_score.info()
```

```python
#change status dtype to int64
cred_score.STATUS=cred_score.STATUS.replace("X","10") # "X" encoded by "10"
cred_score.STATUS=cred_score.STATUS.replace("C","20") # "C" encoded by "20"
cred_score.STATUS= cred_score.STATUS.astype("int64")  # change dtype to itn64
```

```python
#classify customer ID by credit record File.
bad_cus=[] #bad customer ID
good_cus=[] # good customer ID
for i in cred_score.ID.unique():
  a= (cred_score[cred_score.ID==i].STATUS>=2) & (cred_score[cred_score.ID==i].STATUS<=5) 
  if(any(a)):
    bad_cus.append(i)
  else:
    good_cus.append(i)
```

#Pengabungan File dan Pembersihan Duplikat

```python
# create response variable based on previous cell information.
clean_data['CUSTOMER_TYPE']= 'unknown'

for i in bad_cus:
  clean_data.loc[clean_data.ID==i, 'CUSTOMER_TYPE']= 'bad'

for i in good_cus:
  clean_data.loc[clean_data.ID==i, 'CUSTOMER_TYPE']= 'good'
```

```python
clean_data=clean_data[clean_data.CUSTOMER_TYPE != "unknown"] #drop "unknown" row
```

```python
clean_data
```

```python
# Drop Duplicate based on total similarity entry of all columns (except for ID)
clean_data_final=clean_data.drop("ID",axis=1);
clean_data_final=clean_data_final.drop_duplicates()
```

```python
#properly configure dtype for "FLAG" columns
FLAG=["FLAG_MOBIL","FLAG_WORK_PHONE","FLAG_PHONE", "FLAG_EMAIL"]
clean_data_final[FLAG]= clean_data_final[FLAG].astype("object")
```
